```python
#Import required libraries
import pandas as pd
import numpy as np
df = pd.read_csv("/content/3) Sentiment dataset.csv")
```

```python
#-------TEXT PREPROCESSING---------
# For text cleaning
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
#Download required NLTK resources (run once)
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Function to clean and preprocess text
def preprocess_text(text):
    text = str(text).lower()  # Convert to lowercase
    # Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    # Tokenize text into words
    words = word_tokenize(text)
    # Remove stopwords
    words = [word for word in words if word not in stopwords.words('english')]
    # Join words back into sentence
    return " ".join(words)
    # Apply preprocessing to 'Text' column
    df['Text'] = df['Text'].apply(preprocess_text)
```

```python
# ===================== SENTIMENT CONVERSION =====================
# Clean sentiment column
df['Sentiment'] = df['Sentiment'].astype(str).str.strip().str.lower()
# Function to convert sentiment into numerical values
def convert_sentiment(text):
    if 'positive' in text:
        return 1      # Positive
    elif 'negative' in text:
        return -1     # Negative
    else:
        return 0      # Neutral
# Apply conversion
df['Sentiment'] = df['Sentiment'].apply(convert_sentiment)

# Check values
print(df['Sentiment'].unique())

```

```python
#Feature Extraction
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TF-IDF vectorizer
vectorizer = TfidfVectorizer(max_features=5000)

# Convert text data into numerical feature matrix
X = vectorizer.fit_transform(df['Text'])

# Target variable
y = df['Sentiment']
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
# Split data into training (80%) and testing (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#Train Model
from sklearn.linear_model import LogisticRegression

# Initialize Logistic Regression model
model = LogisticRegression()

# Train model using training data
model.fit(X_train, y_train)
```

```python
#prediction
# Predict sentiment for test data
y_pred = model.predict(X_test)
```

```python
#Evaluation
from sklearn.metrics import classification_report

# Print precision, recall, and F1-score
print(classification_report(y_test, y_pred))
```

```python
#Objective: Classify text data into sentiment categories (positive, negative, neutral).

#Dataset: Social media text data with sentiment labels.

#Preprocessing: Cleaned text by converting to lowercase, removing special characters, tokenization, and stopwords removal.

#Feature Extraction: Used TF-IDF to convert text into numerical features.

#Model: Applied Logistic Regression for classification.

#Prediction: Model predicted sentiment labels for test data.

#Evaluation: Performance evaluated using precision, recall, and F1-score.

#Conclusion: The model successfully classified text data into sentiment categories.
```
