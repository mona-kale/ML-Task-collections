```python
#drop 'State' column due to high number of categories
df.drop('State',axis=1,inplace=True)
```

```python
print(df['International plan'].unique())
print(df['Voice mail plan'].unique())
print(df['Churn'].unique())
```

```python
#Handling Categorical Data (encoding)
#convert yes/no value into 1/0
df['International plan'] = df['International plan'].map({'Yes':1, 'No':0})
df['Voice mail plan'] = df['Voice mail plan'].map({'Yes':1, 'No':0})
#convert target variables(True/False into 1/0)
df['Churn'] = df['Churn'].map({True:1, False:0})
```

```python
#Standardization to map all the values on same scale
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
exclude_cols=['Churn','International plan','Voice mail plan']
num_cols= df.drop(exclude_cols,axis=1).select_dtypes(include=['int64','float64']).columns
df[num_cols] = scalar.fit_transform(df[num_cols])
```
